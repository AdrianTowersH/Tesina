```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_bar_chart(df, title):
    # Checamos si el DataFrame tenga las colummnas necesesarias
    if 'Categoría' not in df.columns or 'Tiempo' not in df.columns:
        raise ValueError("DataFrame must contain 'Categoría' and 'Tiempo' columns")

    # Definimos los colores de la grafica
    colors = ['blue', 'orange', 'green']

    # Creamos la grafica de barras
    df.plot(kind='bar', x='Categoría', y='Tiempo', color=colors, legend=False)

    # Mostramos los indicadores del grafico
    plt.title(title)
    plt.xlabel('Categoría de Tiempo')
    plt.ylabel('Tiempo (segundos)')

    # Mostramos la grafica de barras
    plt.show()

# definimos los parametros para cada caso en especifico
data = {
    'Categoría': ['Real', 'User', 'Sys'],
    'Tiempo': [3.1766, 0.2433, 0.0333]
}
df = pd.DataFrame(data)
plot_bar_chart(df, 'Ataque 1 SQLMap: Vista de nombres de las tablas en la  base de datos')
```

```python

def plot_comparation_bar_chart(df, title):

    if 'Herramienta' not in df.columns or 'Tiempo' not in df.columns:
        raise ValueError("DataFrame must contain 'Herramienta' and 'Tiempo' columns")

    colors = ['red', 'blue']

    df.plot(kind='bar', x='Herramienta', y='Tiempo', color=colors, legend=False)

    plt.title(title)
    plt.xlabel('Herramienta')
    plt.ylabel('Tiempo (segundos)')

    plt.show()


data = {
    'Herramienta': ['SQLMap', 'Zap'],
    'Tiempo': [3.2766, 0.03666]
}
df = pd.DataFrame(data)
plot_comparation_bar_chart(df, 'Comparación entre el tiempo REAL de SQLMap Contra OWASP ZAP para el ataque 1')
```
